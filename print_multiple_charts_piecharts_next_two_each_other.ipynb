```python
import matplotlib.pyplot as plt

# Data for the first pie chart
labels = ['Category 1', 'Category 2', 'Category 3']
sizes = [25, 50, 25]

# Data for the second pie chart
labels2 = ['Category A', 'Category B', 'Category C']
sizes2 = [50, 25, 25]



# Create the figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first pie chart
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.set_title('Pie Chart 1')

# Plot the second pie chart
ax2.pie(sizes2, labels=labels2, autopct='%1.1f%%')
ax2.set_title('Pie Chart 2')


# Show the plot
plt.show()
```

```python
#three next to each other lolol
import matplotlib.pyplot as plt

# Data for the first pie chart
labels1 = ['Category 1', 'Category 2', 'Category 3']
sizes1 = [25, 50, 25]

# Data for the second pie chart
labels2 = ['Category A', 'Category B', 'Category C']
sizes2 = [50, 25, 25]

# Data for the third pie chart
labels3 = ['Category X', 'Category Y', 'Category Z']
sizes3 = [30, 40, 30]

# Create the figure and three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(35, 25))

# Plot the first pie chart
ax1.pie(sizes1, labels=labels1, autopct='%1.1f%%')
ax1.set_title('Pie Chart 1')

# Plot the second pie chart
ax2.pie(sizes2, labels=labels2, autopct='%1.1f%%')
ax2.set_title('Pie Chart 2')

# Plot the third pie chart
ax3.pie(sizes3, labels=labels3, autopct='%1.1f%%')
ax3.set_title('Pie Chart 3')

# Show the plot
plt.show()
```
